```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importamos el archivo csv
df = pd.read_csv('Ventas_totales.csv')
df.head()
```

```python
# Identificar valores nulos por columan: Se utilizara para ver con que columnas vamos a trabajar
valores_nulos = df.isnull().sum()
# Hacemos una copia del csv
df1 = df.copy()

valores_nulos
```

# **Sustituir valores nulos con promedio o media**

```python
df1['tarjetas_debito'] = df1['tarjetas_debito'].fillna(round(df['tarjetas_debito'].mean(),1))
df1.to_csv('Columna_tarjetas_debito_un_nulo_metodo_promedio.csv', index=False)
```

```python
df1['otros_medios'] = df1['otros_medios'].fillna(round(df['otros_medios'].mean(),1))
df1.to_csv('Columna_otros_medios_un_nulo_metodo_promedio.csv', index=False)
```

```python
df1['bebidas'] = df1['bebidas'].fillna(round(df['bebidas'].mean(),1))
df1.to_csv('Columna_bebidas_un_nulo_metodo_promedio.csv', index=False)
```

# **Sustitucion de valores nulos con mediana**

```python
df1['subtotal_ventas_alimentos_bebidas'] = df1['subtotal_ventas_alimentos_bebidas'].fillna(round(df['subtotal_ventas_alimentos_bebidas'].median(),1))
df1.to_csv('Columna_subtotal_ventas_alimentos_bebidas_epaciosgrandes_nulo_metodo_mediana.csv', index=False)
```

```python
df1['salon_ventas'] = df1['salon_ventas'].fillna(round(df['salon_ventas'].median(),1))
df1.to_csv('Columna_salon_ventas_epaciosgrandes_nulo_metodo_mediana.csv', index=False)
```

```python
df1['tarjetas_credito'] = df1['tarjetas_credito'].fillna(round(df['tarjetas_credito'].median(),1))
df1.to_csv('Columna_tarjetas_credito_epaciosgrandes_nulo_metodo_mediana.csv', index=False)
```

# **Susutitucion por un valor numerico en concreto.**
Se agarro un numero aleatorio de la columna para insertarlo

```python
df1['indumentaria_calzado_textiles_hogar'] = df1['indumentaria_calzado_textiles_hogar'].fillna(964553.3924)
df1.to_csv('Columna_indumentaria_calzado_textiles_hogar_par_nulos_metodo_numerico.csv', index=False)
```

```python
df1['almacen'] = df1['almacen'].fillna(6705475.318)
df1.to_csv('Columna_almacen_un_nulo_metodo_numerico.csv', index=False)
```

```python
df1['panaderia'] = df1['panaderia'].fillna(4860317.680)
df1.to_csv('Columna_panaderia_un_nulo_metodo_numerico.csv', index=False)
```

# **Sustitucion valores nulos por valores no nulos hacia adelante 'forward fill' ('ffill').**
Hay valores entre nulos uno y otro

```python
df1['otros'] = df1['otros'].fillna(method = 'ffill')
df1.to_csv('Columna_otros_uno_si_uno_no_nulos_metodo_hacia_adelante.csv', index=False)
```

```python
df1['lacteos'] = df1['lacteos'].fillna(method = 'ffill')
df1.to_csv('Columna_lacteos_un_nulo_meodo_hacia_adelante.csv', index=False)
```

```python
df1['carnes'] = df1['carnes'].fillna(method = 'ffill')
df1.to_csv('Columna_carnes_un_nulo_metodo_hacia_adelante.csv', index=False)
```

# **Sustitucion valores nulos por valores no nulos hacia atrás backward fill ("bfill").**
Hay valores nulos entre uno y otro

```python
df1['electronicos_articulos_hogar'] = df1['electronicos_articulos_hogar'].fillna(method = 'bfill')
df1.to_csv('Columna_electronicos_articulos_hogar_uno_si_uno_no_nulos_metodo_hacia_atras.csv', index=False)
```

```python
df1['verduleria_fruteria'] = df1['verduleria_fruteria'].fillna(method = 'bfill')
df1.to_csv('Columna_verduleria_fruteria_un_nulo_metodo_hacia_atras.csv', index=False)
```

```python
df1['alimentos_preparados_rotiseria'] = df1['alimentos_preparados_rotiseria'].fillna(method = 'bfill')
df1.to_csv('Columna_alimentos_preparados_rotiseria_un_nulo_metodo_hacia_atras.csv', index=False)
```

```python
valores_nulos = df1.isnull().sum()
valores_nulos
```
