```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importamos el archivo csv
df = pd.read_csv('Ventas_totales.csv')
df.head()
```

```python
# Identificar valores nulos por columan: Se utilizara para ver con que columnas vamos a trabajar
valores_nulos = df.isnull().sum()
# Hacemos una copia del csv
df1 = df.copy()

valores_nulos
```

```python
# Sustituir valores nulos con promedio o media
#
df1['tarjetas_debito'] = df1['tarjetas_debito'].fillna(round(df['tarjetas_debito'].mean(),1))
df1['otros_medios'] = df1['otros_medios'].fillna(round(df['otros_medios'].mean(),1))
df1['bebidas'] = df1['bebidas'].fillna(round(df['bebidas'].mean(),1))
df1['almacen'] = df1['almacen'].fillna(round(df['almacen'].mean(),1))
df1['panaderia'] = df1['panaderia'].fillna(round(df['panaderia'].mean(),1))
df1['lacteos'] = df1['lacteos'].fillna(round(df['lacteos'].mean(),1))
df1['carnes'] = df1['carnes'].fillna(round(df['carnes'].mean(),1))
df1['verduleria_fruteria'] = df1['verduleria_fruteria'].fillna(round(df['verduleria_fruteria'].mean(),1))
df1['alimentos_preparados_rotiseria'] = df1['alimentos_preparados_rotiseria'].fillna(round(df['alimentos_preparados_rotiseria'].mean(),1))

df1.to_csv('Columnas_un_nulo_metodo_promedio.csv', index=False)

valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
# Sustitucion de valores nulos con mediana
df1['subtotal_ventas_alimentos_bebidas'] = df1['subtotal_ventas_alimentos_bebidas'].fillna(round(df['subtotal_ventas_alimentos_bebidas'].median(),1))
df1['salon_ventas'] = df1['salon_ventas'].fillna(round(df['salon_ventas'].median(),1))
df1['tarjetas_credito'] = df1['tarjetas_credito'].fillna(round(df['tarjetas_credito'].median(),1))

df1.to_csv('Columnas_epaciosgrandes_nulo_metodo_mediana.csv', index=False)

valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
# Susutitucion por un valor numerico en concreto. se agarro un numero aleatorio de la columna para insertarlo
df1['indumentaria_calzado_textiles_hogar'] = df1['indumentaria_calzado_textiles_hogar'].fillna(964553.3924)

df1.to_csv('Columnas_par_nulos_metodo_numerico.csv', index=False)

valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
# Sustitucion valores nulos por valores no nulos hacia adelante 'forward fill' ('ffill'), hay valores entre nulos uno y otro
df1['otros'] = df1['otros'].fillna(method = 'ffill')

df1.to_csv('Columnas_uno_si_uno_no_nulos_hacia_adelante.csv', index=False)

valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
# Sustitucion valores nulos por valores no nulos hacia atrás backward fill ("bfill"). hay valores nulos entre uno y otro
df1['electronicos_articulos_hogar'] = df1['electronicos_articulos_hogar'].fillna(method = 'ffill')

df1.to_csv('Columnas_uno_si_uno_no_nulos_hacia_atras.csv', index=False)

valores_nulos = df1.isnull().sum()
valores_nulos
```
