```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('listings.csv')
df1 = df.copy()
```

```python
nulos = df.isnull().sum()
nulos
```

```python
#Sustituir valores nulos con promedio o media
df1['review_scores_rating'] = df1['review_scores_rating'].fillna(round(df['review_scores_rating'].mean(),1))
df1['review_scores_accuracy'] = df1['review_scores_accuracy'].fillna(round(df['review_scores_accuracy'].mean(),1))
df1['review_scores_cleanliness'] = df1['review_scores_cleanliness'].fillna(round(df['review_scores_cleanliness'].mean(),1))
df1['review_scores_checkin'] = df1['review_scores_checkin'].fillna(round(df['review_scores_checkin'].mean(),1))
df1['review_scores_communication'] = df1['review_scores_communication'].fillna(round(df['review_scores_communication'].mean(),1))
df1['review_scores_location'] = df1['review_scores_location'].fillna(round(df['review_scores_location'].mean(),1))
df1['review_scores_value'] = df1['review_scores_value'].fillna(round(df['review_scores_value'].mean(),1))
```

```python
# Sustituir valores nulos con mediana
df1['host_response_rate'] = df1['host_response_rate'].fillna(round(df['host_response_rate'].median(),1))
df1['host_acceptance_rate'] = df1['host_acceptance_rate'].fillna(round(df['host_acceptance_rate'].median(),1))
df1['host_listings_count'] = df1['host_listings_count'].fillna(round(df['host_listings_count'].median(),1))
df1['host_total_listings_count'] = df1['host_total_listings_count'].fillna(round(df['host_total_listings_count'].median(),1))
```

```python
# Susutitucion por un valor numerico en concreto
# df1['description'] = df1['description'].fillna()
```

```python
# Sustituir valores nulos por un string en concreto
df1['description'] = df1['description'].fillna('Sin descripción')
df1['neighborhood_overview'] = df1['neighborhood_overview'].fillna('Sin descripción')
df1['host_is_superhost'] = df1['host_is_superhost'].fillna('f')
df1['host_location'] = df1['host_location'].fillna('f')
df1['has_availability'] = df1['has_availability'].fillna('t')
df1['first_review'] = df1['first_review'].fillna('Sin fecha')
df1['last_review'] = df1['last_review'].fillna('Sin fecha')
```

```python
# sustituir valores nulos por valores no nulos hacia adelante 'forward fill' ('ffill')
df1['bathrooms_text'] = df1['bathrooms_text'].fillna(method = 'ffill')
df1['bedrooms'] = df1['bedrooms'].fillna(method = 'ffill')
df1['beds'] = df1['beds'].fillna(method = 'ffill')
```

```python
# Sustituir valores nulos por valores no nulos hacia atrás backward fill ("bfill")
df1['host_response_time'] = df1['host_response_time'].fillna(method = "bfill")
df1['host_neighbourhood'] = df1['host_neighbourhood'].fillna(method = "bfill")
df1['neighbourhood'] = df1['neighbourhood'].fillna(method = "bfill")
df1['host_verifications'] = df1['host_verifications'].fillna(method = "bfill")
df1['host_identity_verified'] = df1['host_identity_verified'].fillna(method = "bfill")
```

```python
nulos = df1.isnull().sum()
nulos
```

```python
# Eliminamos filas con nulos
df1 = df1.dropna()

df1.to_csv('listings_limpio.csv', index=False)
```
